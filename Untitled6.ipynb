```python
!apt-get install tesseract-ocr -y
!pip install pytesseract --quiet

import requests, json, re, cv2, pytesseract
import numpy as np
import matplotlib.pyplot as plt
from ast import literal_eval

pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

print("🧠 Enter your app prompt (e.g., 'Create a 4-screen productivity app'):")
user_prompt = input()

DEEPINFRA_API_KEY = "9bHbBBEfkKtr01NeJPIpAKG6YcfsvUhX"
API_URL = "https://api.deepinfra.com/v1/openai/chat/completions"
headers = {
    "Authorization": f"Bearer {DEEPINFRA_API_KEY}",
    "Content-Type": "application/json"
}

system_prompt = """
You are an AI UX assistant. The user gives a prompt for an app (e.g., productivity, education).
Return JSON defining multiple mobile screens. Each screen must be a list of UI components.
Each component must have:
- "type" (e.g. text, button, image, input)
- "label" (short text)
- "position": [x, y]
- "size": [width, height]

Return JSON like:
{
  "screen1": [
    {"type": "text", "label": "Welcome", "position": [50, 40], "size": [300, 40]}
  ],
  "screen2": [
    ...
  ]
}
Return only valid JSON. No explanation.
"""

payload = {
    "model": "meta-llama/Meta-Llama-3-8B-Instruct",
    "messages": [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    "temperature": 0.3
}

response = requests.post(API_URL, headers=headers, json=payload)
result = response.json()
ai_reply = result["choices"][0]["message"]["content"]
print("🔍 Raw LLM output:\n", ai_reply)

match = re.search(r'(\{.*\})', ai_reply, re.DOTALL)
if not match:
    raise ValueError("Could not extract valid layout.")

fixed_json = re.sub(r'\}\s*\{', '}, {', match.group(1))

try:
    layout_dict = json.loads(fixed_json)
except json.JSONDecodeError:
    layout_dict = literal_eval(fixed_json)

# Render each screen
for screen_name, elements in layout_dict.items():
    canvas = np.ones((600, 350, 3), dtype=np.uint8) * 255

    for el in elements:
        if "position" in el and "size" in el:
            x, y = el["position"]
            w, h = el["size"]
            label = el.get("label", el["type"])
            cv2.rectangle(canvas, (x, y), (x + w, y + h), (0, 0, 0), 2)
            cv2.putText(canvas, label, (x + 5, y + h//2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (20, 20, 20), 1)

    plt.figure(figsize=(4, 8))
    plt.imshow(canvas)
    plt.axis('off')
    plt.title(screen_name)
    plt.show()
```
