```python
from groq import Groq

client = Groq()
completion = client.chat.completions.create(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    messages=[
        {
            "role": "user",
            "content": "Web App Name: SmartClass CRM Purpose: The primary purpose of this web app is to help schools manage their academic operations — including student records, class schedules, attendance, grading, and communication between teachers and parents. Target Audience: The app is designed for school administrators, teachers, and parents — mainly adults aged 25–50 working in education. Key Features: Login/signup functionality, a teacher dashboard with quick access to daily tasks, a searchable student list, a digital grade journal, attendance tracking, messaging between teachers and parents, and a notification center for important updates. Color Scheme: I’d like a professional and calm theme using blue (#1E3A8A), white, and light gray. Style: The design should be modern and minimalist, with clean lines, readable fonts, and mobile responsiveness. Number of Pages: I’d like 6 pages in total — a homepage/login screen, the main teacher dashboard, a journal entry page, student profile pages, a notification view, and a settings page."
        }
    ],
    temperature=1,
    max_completion_tokens=1024,
    top_p=1,
    stream=True,
    stop=None,
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```
