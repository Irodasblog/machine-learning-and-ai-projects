```python
df = df[['district', 'rooms', 'size', 'level', 'max_levels', 'price', 'lat', 'lng']]
```

```python
df = pd.get_dummies(df, columns=['district'])
```

```python
X = df.drop(columns=['price'])
y = df['price']
```

```python
y_log = np.log1p(y)
```

```python
X_train, X_test, y_train_log, y_test_log = train_test_split(X, y_log, test_size=0.2, random_state=42)
```

```python
X_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X)
```

```python
y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=X_scaled.shape[1]))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_absolute_error')
```

```python
model.fit(X_scaled, y_scaled, epochs=200, batch_size=32, validation_split=0.2, verbose=1)
```

```python
predictions_scaled = model.predict(X_scaled)
predictions = y_scaler.inverse_transform(predictions_scaled)
```

```python
mae = mean_absolute_error(y, predictions)
r2 = r2_score(y, predictions)
print(f"RВІ (accuracy): {r2 * 100:.2f}%")
```
