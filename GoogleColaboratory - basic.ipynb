Functions


Functions is a block of code that can be reused in a Program

```python
#Factorial of a Number
#is the product of alll the positive integres less
# than or equal to the given number

# Factorial of 5 = 5 * 4 * 3 * 2 * 1 = 120
number  = int(input('Enetr a number to find its factorial: '))

factorial = 1
if number == 0:
  print('The factorial of 0 is 1 ')
else:
  for i in range(1, number+1): # range(1,6) ---> 1,2,3,4,5
    factorial = factorial*i #1*2*3*4*5
  print('The factorial of ', number, 'is' , factorial)

```

```python
#factorial function
def factorial_value(num):

  factorial = 1

  if num == 0:
    return factorial
  else:
    for i in range(1, number+1):
      factorial = factorial*i
    return factorial



```

```python
print(factorial_value(10))
```
