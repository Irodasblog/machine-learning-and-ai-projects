```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_parquet("train-00000-of-00003.parquet")
df = df.sample(frac=0.5, random_state=42)

texts = df['text'].tolist()
categories = df['folder'].tolist()

print(f"Using {len(texts)} samples.")
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

def predict_category(input_text):
    input_vec = vectorizer.transform([input_text])
    similarity_scores = cosine_similarity(input_vec, X)

    best_match_index = similarity_scores.argmax()

    predicted_category = categories[best_match_index]
    matched_text = texts[best_match_index]

    return {
        "Input Text": input_text,
        "Matched Example": matched_text,
        "Predicted Category": predicted_category
    }

user_input_text = input("Enter your Uzbek news text: ")
result = predict_category(user_input_text)

print("\nPrediction Result:")
for key, value in result.items():
    print(f"{key}: {value}")

```
